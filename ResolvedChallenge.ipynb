# PyCity Schools Analysis


The dataset provides comprehensive information on 15 distinct schools, consisting of the following key statistics:

-School Count: There are a total of 15 schools in the dataset.
-Student Count: The dataset encompasses 39,170 students in total.
-Total Budget: The cumulative budget allocated to these 15 schools amounts to $24,649,428.00.
-Average Math Score: The average math score across all students in the dataset is 78.99.
-Average Reading Score: The average reading score among all students in the dataset is 81.88.
-Approximately 74.98 % of students achieved a passing score in mathematics.
-A substantial majority of students, around 85.81 %, successfully passed their reading assessments.
-The overall passing rate, considering both math and reading, is calculated to be 65.17%.

Our primary objective was to explore the connection between the financial resources allocated to each educational institution and their impact on the quality of education. This quality was measured by assessing student performance in mathematics, reading, and the combined performance in both subjects.
Our analysis has yielded several key findings:

Charter Schools Outperform District Schools: The data unequivocally illustrates that Charter schools consistently outperform District schools in terms of student achievement. This conclusion holds true despite the fact that Charter schools often operate with a lower average budget than their District counterparts.

Per Student Budget's Limited Impact: Surprisingly, our analysis reveals that the per-student budget does not appear to be a critical factor influencing student performance. Charter schools, with their lower budgets, are still able to achieve better results than District schools.

Consistency Across Grade Levels: Our examination of performance across different grade levels within the sample showed remarkable consistency. This suggests that the performance trends observed are representative of schools in general, and there are no significant variations based on students' grade levels.

Size Matters in Performance: An in-depth analysis of school performance by size reveals an intriguing pattern. Medium-sized schools consistently emerge as the top performers, achieving higher overall qualifications. Conversely, larger schools consistently perform less favorably, demonstrating a need for further investigation and targeted interventions in these institutions.

In summary, this study underscores the influence of school type and size on student performance. It emphasizes the potential for Charter schools to achieve superior results, irrespective of budget constraints, and highlights the pivotal role of school size in shaping academic outcomes. These insights can inform educational policy and decision-making to enhance the overall quality of education in these institutions

One of the key limitations of our analysis lies in the absence of detailed information about the specific criteria that define the budget allocations for each educational institution. Without insights into the factors that shape these budgets, such as funding sources, allocation methodologies, or budget utilization, we face constraints in fully understanding the financial dynamics of each school.

Furthermore, our analysis lacks comprehensive data regarding potential differences in the curriculum, educational approaches, or teaching styles between Charter and District schools. Such information could provide valuable context for interpreting the performance disparities observed in our study. The absence of these details restricts our ability to explore the underlying reasons for these disparities fully.

Additionally, we do not possess data concerning the allocation of instructional time to different subjects within both Charter and District schools. Variations in the allocation of time to subjects like math and reading may play a significant role in the observed differences in student performance. This absence of data related to instructional time represents a limitation in our analysis

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete['school_name'].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete['Student ID'].nunique()
student_count
```

```python
# Calculate the total budget
total_budget = school_data['budget'].sum()

total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete ['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete ['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"] 
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_data = [
    ['School Count', school_count],
    ['Student Count', student_count],
    ['Total Budget', "${:,.2f}".format(total_budget)],
    ['Average Math Score','{:,.2f}'.format(average_math_score)],
    ['Average Reading Score', '{:,.2f}'.format(average_reading_score)],
    ['Passing Math Percentage', '{:,.2f}'.format(passing_math_percentage)],
    ['Passing Reading Percentage', '{:,.2f}'.format(passing_reading_percentage)],
    ['Overall Passing Rate', '{:,.2f}'.format(overall_passing_rate)]
]

# Create a DataFrame from the list of lists
district_summary_df = pd.DataFrame(district_summary_data) 

# Display the DataFrame
print(district_summary_df)
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types_f = school_data_complete[['school_name', 'type']].drop_duplicates().set_index('school_name')

school_types=pd.DataFrame(school_types_f )
school_types
```

```python
# Calculate the total student count per school
per_school_counts_f = school_data_complete.groupby('school_name')['Student ID'].nunique()

per_school_counts=pd.DataFrame(per_school_counts_f)
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget_f = school_data.groupby('school_name')['budget'].sum()
per_school_capita_f= per_school_budget_f/per_school_counts_f

per_school_budget=pd.DataFrame({'Per School Budget': per_school_budget_f.map("${:,.0f}".format)})
per_school_capita=pd.DataFrame({'Per School Capita':per_school_capita_f})
 
per_school_money = pd.concat([per_school_budget, per_school_capita], axis=1)    

per_school_money
```

```python
# Calculate the average test scores per school
per_school_math_f = school_data_complete.groupby('school_name')['math_score'].mean()
per_school_reading_f= school_data_complete.groupby('school_name')['reading_score'].mean()

per_school_math=pd.DataFrame({'Per School Math':per_school_math_f})
per_school_reading=pd.DataFrame({'Per School Reading': per_school_reading_f})
 
per_school_math_and_reading = pd.concat([per_school_math, per_school_reading], axis=1) 
per_school_math_and_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete['math_score']>=70].groupby('school_name')['Student ID'].count()
school_students_passing_math = pd.DataFrame({'Students per School Passed Math':students_passing_math})
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete['reading_score']>=70].groupby('school_name')['Student ID'].count()
school_students_passing_reading = pd.DataFrame({'Students per School Pass Reading':students_passing_reading})
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading_f = students_passing_math_and_reading.groupby(["school_name"]).size()

school_students_passing_math_and_reading=pd.DataFrame({'Students Pass Math & Read': school_students_passing_math_and_reading_f})

school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = students_passing_math / (per_school_counts_f) * 100
per_school_passing_reading = students_passing_reading/ (per_school_counts_f) * 100
overall_passing_rate = school_students_passing_math_and_reading_f/ (per_school_counts_f) * 100

per_school_passing_math=pd.DataFrame({'%Passing Math':per_school_passing_math}) 
per_school_passing_reading =pd.DataFrame({'%Passing Reading':per_school_passing_reading})
overall_passing_rate=pd.DataFrame({'%Passing Overall':overall_passing_rate})


passing_rates=pd.concat([per_school_passing_math,per_school_passing_reading,overall_passing_rate],axis=1)

print(passing_rates)

```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.concat([school_types, per_school_counts,per_school_money, per_school_math_and_reading,
                                school_students_passing_math,school_students_passing_reading,
                                school_students_passing_math_and_reading,
                                school_students_passing_math_and_reading,passing_rates], axis=1)

# Formatting

# Display the DataFrame
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values(by=['%Passing Overall'],ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary.sort_values(by=['%Passing Overall'],ascending=True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby('school_name')['math_score'].mean()
tenth_grader_math_scores = tenth_graders.groupby('school_name')['math_score'].mean()
eleventh_grader_math_scores = eleventh_graders.groupby('school_name')['math_score'].mean()
twelfth_grader_math_scores =twelfth_graders.groupby('school_name')['math_score'].mean() 

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({'9th':ninth_grade_math_scores.map("{:,.2f}".format), 
                                     '10th':tenth_grader_math_scores.map("{:,.2f}".format),
                           '11th':eleventh_grader_math_scores.map("{:,.2f}".format),
                            '12th':twelfth_grader_math_scores.map("{:,.2f}".format) })

# Minor data wrangling
math_scores_by_grade = math_scores_by_grade[["9th", "10th", "11th", "12th"]] 
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby('school_name')['reading_score'].mean()
tenth_grader_reading_scores = tenth_graders.groupby('school_name')['reading_score'].mean()
eleventh_grader_reading_scores = eleventh_graders.groupby('school_name')['reading_score'].mean()
twelfth_grader_reading_scores =twelfth_graders.groupby('school_name')['reading_score'].mean() 

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({'9th':ninth_grade_reading_scores.map("{:,.2f}".format), 
                                        '10th':tenth_grader_reading_scores.map("{:,.2f}".format),
                           '11th':eleventh_grader_reading_scores.map("{:,.2f}".format),
                            '12th':twelfth_grader_reading_scores.map("{:,.2f}".format) })

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Per School Capita"] = school_spending_df["Per School Capita"].astype(float)

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per School Capita'], spending_bins, labels=labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])['Per School Math'].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])['Per School Reading'].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["%Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["%Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])['%Passing Overall'].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({'Spending Math Scores':spending_math_scores,'Spending Reading Scores':spending_reading_scores,
                                 'Spending %Math':spending_passing_math,'Spending %Reading': spending_passing_reading,
                                 'Spending % Overall':overall_passing_spending}) 

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary['Student ID'], size_bins, labels=labels)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])['Per School Math'].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])['Per School Reading'].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["%Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["%Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])['%Passing Overall'].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({'Size Math Scores':size_math_scores,'Size Reading Scores':size_reading_scores,
                'Size Passing Math':size_passing_math, 'Size Passing Reading': size_passing_reading,
                'Size Overall Passing':size_overall_passing})

# Display results
size_summary
```
